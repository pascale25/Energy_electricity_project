# KOUMASSI_AT1

```python
# remane all columns
def renamecolumns(df):
  df.rename(columns={'Jours': 'date', 'Heures':'hours', 'P(kW)':'P', 'Q(kVAR)':'Q', 'S(kVA)':'S', 'Cos Φ': 'Cos', 'charge':'charge',
       'nom':'name', 'puissance':'power'}, inplace=True)
  return df

def drop_repalce_nan(df):
  print(df.isnull().sum())
  df.dropna(subset = ['P', 'Q','S'], inplace = True)
  df['Cos']= df['Cos'].fillna(0)
  return df

def charge_cos(df):
  df['Cos']=df['P']/df['S']
  df['charge']=df['S']/df['power']
  return df 

def convert_date(df):
  #df['TEMPS'] =  pd.to_datetime(df21['TEMPS'], format='%d.%m.%Y %H:%M:%S')
  df['hours'] =  pd.to_datetime(df['hours'], format= '%H:%M:%S')
  #df['date'] = df['date'].dt.date
  df['year'] = df['date'].dt.year
  df['month'] = df['date'].dt.month
  df['day'] = df['date'].dt.day
  df['hour'] = df['hours'].dt.hour
  df['weekday'] = df['date'].dt.weekday
  #df['day_name'] = df['date'].dt.weekday_name
  #df['dayofyear'] = df['date'].dt.dayofyear
  return df
```
